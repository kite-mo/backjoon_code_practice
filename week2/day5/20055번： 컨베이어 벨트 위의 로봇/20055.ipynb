```python
import sys
from collections import deque
import time 

sys.stdin = open('./test.txt')
N, K = list(map(int, sys.stdin.readline().strip().split()))
durability_queue = deque(map(int, sys.stdin.readline().strip().split()))
robot_queue = deque([False] * (2*N))
count_k = 0
count = 0

drop_robot_index = N-1
belt_keep_moving = True

while belt_keep_moving:
    count += 1
    # first : belt move
    durability_queue.rotate()
    robot_queue.rotate()

    # drop robot
    robot_queue[drop_robot_index] = False

    # second : move robot
    for robot_index in range(N-2, -1, -1):
        # 조건문 실행 시, 변수 정의를 하지 않고 바로 조건에 넣는게 실행 시간 측면에서 유리함
        if robot_queue[robot_index] and not robot_queue[robot_index+1] and durability_queue[robot_index+1] > 0:
            robot_queue[robot_index+1] = True
            robot_queue[robot_index] = False
            durability_queue[robot_index+1] -= 1
            if durability_queue[robot_index+1] == 0:
                count_k += 1
    
    # drop robot
    robot_queue[drop_robot_index] = False

    # third : add robot first index
    if durability_queue[0] > 0:
        robot_queue[0] = True
        durability_queue[0] -= 1
        if durability_queue[0] == 0:
            count_k += 1

    # fourth
    if count_k >= K:
        break

print(count)
```
